#Import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#1. Data Loading and Overview

```python
# Load the Iris dataset using seaborn
iris = sns.load_dataset('iris')

# Display the first few rows
print("First few rows of the Iris dataset:")
print(iris.head())


```

Insights

When I explored the Iris dataset, I noticed it contains measurements of sepal and petal lengths and widths for three flower species: Setosa, Versicolor, and Virginica. This dataset is well-structured and suitable for applying sampling methods and analyzing the Central Limit Theorem.



# 2. Simple Random Sampling

```python

# Randomly sample 30 observations from the dataset
random_sample = iris.sample(n=30, random_state=42)
print("\nSimple Random Sample of 30 observations:")
print(random_sample)

```

Insights

By performing simple random sampling, I randomly selected 30 observations from the dataset. This method ensured that every record had an equal probability of being included, eliminating selection bias. The sample captured a diverse subset of the dataset, representing the population effectively.

# 3. Sample Mean Distribution Analysis

```python
sample_means = []
for _ in range(100):
    sample = iris.sample(n=30, random_state=np.random.randint(0, 1000))
    sample_mean = sample['sepal_length'].mean()
    sample_means.append(sample_mean)
```

```python

# Plot the distribution of the sample means
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sample Means (Sepal Length)')
plt.xlabel('Mean Sepal Length')
plt.ylabel('Frequency')
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=1, label='Mean of Sample Means')
plt.legend()
plt.show()

```

Insights

To validate the Central Limit Theorem, I repeated random sampling 100 times, calculating the mean sepal length for each sample. When I plotted the distribution of these sample means, it resembled a normal distribution, demonstrating that even if the underlying data isnâ€™t perfectly normal, the means of multiple samples converge toward normality.
Moreover, the mean of the sample means (indicated by the red dashed line) was close to the actual population mean of sepal length, proving that repeated sampling provides a reliable estimate of the population mean.

# 4. Systematic Sampling

```python
# Perform systematic sampling with a step size of 10
step = 10
systematic_sample = iris.iloc[::step, :]
print("\nSystematic Sample (every 10th observation):")
print(systematic_sample)

# Visualize the systematic sample distribution for sepal length
plt.figure(figsize=(8, 5))
sns.histplot(systematic_sample['sepal_length'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Systematic Sample Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

Insights

For systematic sampling, I selected every 10th observation from the dataset. This approach was straightforward and ensured a consistent sampling method. However, I observed that systematic sampling could potentially miss some patterns or nuances in the data, especially if there's an inherent periodic trend. Nonetheless, it provided an evenly spaced subset for analysis, useful in situations where simplicity and consistency are prioritized.
