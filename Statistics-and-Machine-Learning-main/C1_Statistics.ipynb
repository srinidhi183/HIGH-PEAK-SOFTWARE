# **Application of Statistics**

## Clinical Trials
In clinical trials, **hypothesis testing** is often used to test the efficacy of a new drug. The plot above shows the distribution of patient responses from two groups: Placebo and Drug. A t-test is performed, and if the p-value is small enough, we reject the null hypothesis that the drug has no effect.


````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Data for drug efficacy trial (Placebo and Drug groups)
np.random.seed(42)
placebo = np.random.normal(50, 10, 100)
drug = np.random.normal(55, 10, 100)

# Perform t-test (hypothesis testing)
t_stat, p_value = stats.ttest_ind(placebo, drug)

# Plot histograms of the results
plt.hist(placebo, bins=15, alpha=0.7, label='Placebo', color='skyblue')
plt.hist(drug, bins=15, alpha=0.7, label='Drug', color='lightgreen')
plt.title(f"Clinical Trial Results")
plt.xlabel("Response")
plt.ylabel("Frequency")
plt.legend()
plt.show()

````

## Blood Pressure Comparison Between Two Treatment Groups

In healthcare, comparing the effects of two treatments on a variable like blood pressure is a common statistical task. A **t-test** is often used to determine whether there is a statistically significant difference between two groups.

In the example above, we compare blood pressure between patients in two treatment groups: Treatment A and Treatment B. The **p-value** from the t-test helps us determine if the difference is statistically significant. A p-value less than 0.05 indicates that the difference in means is significant.

The plot helps healthcare professionals and researchers visually compare treatment outcomes, guiding data-driven decisions on the efficacy of different treatments through statistical significance.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated data for blood pressure in two groups (treatment A and treatment B)
np.random.seed(42)
treatment_A = np.random.normal(120, 10, 50)  # Mean 120, Std Dev 10, 50 patients
treatment_B = np.random.normal(115, 12, 50)  # Mean 115, Std Dev 12, 50 patients

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(treatment_A, treatment_B)

# Plot the distributions of both groups
plt.hist(treatment_A, bins=15, alpha=0.7, label='Treatment A', color='blue')
plt.hist(treatment_B, bins=15, alpha=0.7, label='Treatment B', color='green')
plt.title(f"Blood Pressure Comparison Between Two Treatments (p-value: {p_value:.4f})")
plt.xlabel("Blood Pressure (mmHg)")
plt.ylabel("Number of Patients")
plt.legend()
plt.show()

print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

# Interpretation of the p-value
if p_value < 0.05:
    print("The difference in blood pressure between Treatment A and Treatment B is statistically significant.")
else:
    print("No significant difference in blood pressure between Treatment A and Treatment B.")
````

**Introduction to Infrared (IR) Spectroscopy**
Infrared (IR) Spectroscopy is a vital analytical technique used to identify substances and investigate chemical compositions based on how molecules absorb infrared light. Different molecular bonds absorb characteristic wavelengths of infrared radiation, creating unique spectral fingerprints. This method is widely used for qualitative analysis in various fields such as chemistry, material science, and pharmaceuticals.

**Benefits of Synthetic Data Generation for IR Spectroscopy**
Generating synthetic IR spectroscopy data is beneficial for testing and developing analytical algorithms, training machine learning models, and educational purposes when real data is unavailable or expensive to obtain. The code provided simulates a realistic IR spectrum using Gaussian functions to represent absorption peaks and includes noise to mimic real measurement conditions. By studying these synthetic spectra, researchers can enhance their understanding and interpretation skills without the immediate need for laboratory resources.

````python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, amplitude, mean, stddev):
    """Returns a Gaussian (normal distribution) curve."""
    return amplitude * np.exp(-((x - mean) ** 2) / (2 * stddev ** 2))

# Parameters for synthetic data generation
wavenumbers = np.linspace(400, 4000, 3600)  # Simulated range of wavenumbers from 400 to 4000 cm^-1
peaks = [
    {'amplitude': 1.0, 'mean': 750, 'stddev': 10},   # Example peak at 750 cm^-1
    {'amplitude': 0.7, 'mean': 1500, 'stddev': 20},  # Example peak at 1500 cm^-1
    {'amplitude': 0.5, 'mean': 3000, 'stddev': 15},  # Example peak at 3000 cm^-1
]

# Generate synthetic spectrum by adding Gaussian peaks
spectrum = np.zeros_like(wavenumbers)
for peak in peaks:
    spectrum += gaussian(wavenumbers, **peak)

# Adding random noise to the spectrum
noise = np.random.normal(0, 0.05, spectrum.shape)
spectrum += noise

# Plotting the synthetic spectrum
plt.figure(figsize=(10, 6))
plt.plot(wavenumbers, spectrum, label='Synthetic IR Spectrum')
plt.title('Synthetic IR Spectrum with Gaussian Peaks')
plt.xlabel('Wavenumber (cm^-1)')
plt.ylabel('Absorbance')
plt.legend()
plt.grid(True)
plt.show()
````

**Analyzing and Visualizing Central Tendencies in Python**
This Python example demonstrates how to compute and visualize the central tendenciesmean, median, and modeof a dataset using the libraries NumPy, pandas, and matplotlib. We begin by creating an array of numeric values. Using NumPy, we calculate the mean and median, and with SciPy's stats module, we determine the mode.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Create a sample dataset
data = np.array([23, 24, 29, 20, 32, 25, 25, 30, 24, 35, 24, 25])

# Calculate measures of central tendency
mean_value = np.mean(data)
median_value = np.median(data)
mode_result = stats.mode(data)
mode_value = mode_result.mode  # Access the mode value directly

# Print the results
print("Mean: ", mean_value)
print("Median: ", median_value)
print("Mode: ", mode_value)

# Creating a DataFrame for easier visualization
df = pd.DataFrame(data, columns=['Data'])

# Plotting the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=range(20, 40), color='skyblue', alpha=0.7, label='Data', edgecolor='black')
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_value}')
plt.axvline(mode_value, color='blue', linestyle='dashed', linewidth=1, label=f'Mode: {mode_value}')

plt.title('Histogram of Data with Mean, Median, and Mode')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

**Analyzing and Visualizing Measures of Dispersion in Python**
This Python example illustrates how to compute the measures of dispersionvariance, standard deviation, and rangefor a dataset using the NumPy library. We begin by generating a set of numeric values which serves as our data. Using NumPy, we calculate the variance to gauge the average squared deviations from the mean, the standard deviation to measure the average distance of data points from the mean, and the range to determine the spread between the smallest and largest values. To visually represent these dispersion metrics, we plot a histogram and overlay lines indicating the mean and standard deviations. This visualization is instrumental in understanding the spread and distribution of the data, providing insights into the variability of the dataset.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset
data = np.array([23, 24, 29, 20, 32, 25, 25, 30, 24, 35, 24, 25])

# Calculate measures of dispersion
variance = np.var(data)
standard_deviation = np.sqrt(variance)
data_range = np.max(data) - np.min(data)

# Print the results
print("Variance: ", variance)
print("Standard Deviation: ", standard_deviation)
print("Range: ", data_range)

# Creating a DataFrame for easier visualization
df = pd.DataFrame(data, columns=['Data'])

# Plotting the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=range(20, 40), color='skyblue', alpha=0.7, label='Data', edgecolor='black')
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {np.mean(data):.2f}')
plt.axvline(np.mean(data) - standard_deviation, color='purple', linestyle='dashed', linewidth=1, label=f'-1 SD')
plt.axvline(np.mean(data) + standard_deviation, color='purple', linestyle='dashed', linewidth=1, label=f'+1 SD')

plt.title('Histogram of Data with Variance, Standard Deviation, and Range')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

## Variance and Standard Deviation in Portfolio Management

### Overview:
This notebook demonstrates the application of variance and standard deviation to assess the risk associated with stock investments. We use Python to calculate these metrics for two stocks based on their monthly returns.

### Purpose:
- **Variance** and **Standard Deviation** are key statistical tools used in finance to measure stock return volatility, which indicates risk level.
- Higher values suggest greater risk, which can influence investment decisions.

### Method:
1. **Data Setup**: Two stocks with six months of return data.
2. **Calculations**:
   - Compute the mean, variance, and standard deviation for each stock.
   - Analyze these metrics to compare the investment risks.

### Conclusion:
Understanding these statistics helps investors manage portfolio risk effectively, guiding decisions on which stocks might best fit their risk tolerance and investment goals.


````python
import numpy as np

# Monthly returns (in percentage)
returns_a = np.array([7, 8, 5, 10, 6, 7])
returns_b = np.array([-2, 3, 10, 5, 0, 2])

# Calculating the mean
mean_a = np.mean(returns_a)
mean_b = np.mean(returns_b)

# Variance and Standard Deviation
variance_a = np.var(returns_a)
variance_b = np.var(returns_b)
std_dev_a = np.sqrt(variance_a)
std_dev_b = np.sqrt(variance_b)

print(f"Stock A - Mean: {mean_a:.2f}%, Variance: {variance_a:.2f}, Standard Deviation: {std_dev_a:.2f}")
print(f"Stock B - Mean: {mean_b:.2f}%, Variance: {variance_b:.2f}, Standard Deviation: {std_dev_b:.2f}")
````

## Analysis of Team Score Variability

### Overview:
This notebook demonstrates how a higher standard deviation in game scores might indicate a sports team's potential to perform exceptionally well under certain conditions.

## Properties of Probability Distributions

This notebook delves into the properties of two key probability distributions: Normal and Exponentially Skewed. By generating synthetic data and computing essential statistics, we visualize and analyze how these distributions manifest different properties.

### Overview of Properties Analyzed:
- **Central Tendency**: Mean and median are calculated to indicate the central point of the distributions.
- **Dispersion**: Variance and standard deviation are determined to measure the spread of the data around the central value.
- **Shape**: Skewness and kurtosis are examined to assess the asymmetry and tail thickness of the distributions.

### Visualization Strategy:
- **Histograms and KDE Plots**: Displayed using Seaborn to represent the frequency and density of the data.
- **Annotations**: Skewness and kurtosis values are annotated on the plots to highlight the shape characteristics directly on the visual representation.

### Significance:
- **Normal Distribution**: Ideal for understanding symmetrical data behavior, commonly assumed in many statistical tests.
- **Exponentially Skewed Distribution**: Useful for studying cases with potential extreme values, often seen in risk analysis and queuing theory.

This focused analysis helps in comprehending how different distributions can be used to model real-world data, reflecting on their central tendencies, dispersions, and overall shapes.


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Generate synthetic data
data_normal = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution
data_skewed = np.random.exponential(scale=1, size=1000)    # Skewed distribution

# Calculate statistics
stats_normal = {
    'Mean': np.mean(data_normal),
    'Median': np.median(data_normal),
    'Mode': data_normal[0],  # Note: Mode here is illustrative, typically mode is not just the first element
    'Variance': np.var(data_normal),
    'Std. Deviation': np.std(data_normal),
    'Skewness': skew(data_normal),
    'Kurtosis': kurtosis(data_normal)
}

stats_skewed = {
    'Mean': np.mean(data_skewed),
    'Median': np.median(data_skewed),
    'Mode': data_skewed[0],  # Note: Mode here is illustrative
    'Variance': np.var(data_skewed),
    'Std. Deviation': np.std(data_skewed),
    'Skewness': skew(data_skewed),
    'Kurtosis': kurtosis(data_skewed)
}

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 7), sharey=True)

sns.histplot(data_normal, kde=True, color="blue", ax=axes[0])
axes[0].set_title('Normal Distribution')
axes[0].annotate(f"Skewness: {stats_normal['Skewness']:.2f}\nKurtosis: {stats_normal['Kurtosis']:.2f}",
                 xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10,
                 horizontalalignment='left', verticalalignment='top')

sns.histplot(data_skewed, kde=True, color="green", ax=axes[1])
axes[1].set_title('Exponentially Skewed Distribution')
axes[1].annotate(f"Skewness: {stats_skewed['Skewness']:.2f}\nKurtosis: {stats_skewed['Kurtosis']:.2f}",
                 xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10,
                 horizontalalignment='left', verticalalignment='top')

plt.tight_layout()
plt.show()

# Print statistics
print("Statistics for Normal Distribution:", stats_normal)
print("Statistics for Skewed Distribution:", stats_skewed)
````

# **Sampling Techniques in Statistics**

This notebook demonstrates four common sampling techniques used in statistics and data analysis to gather information from a population without having to survey every individual. Below is a brief description of each sampling method, followed by visual representations of the samples taken using each method.

## **Population**
For demonstration purposes, we define a population of 100 individuals, numbered from 1 to 100. This population represents a city's population or any group of interest.

## **1. Random Sampling**
In random sampling, every individual in the population has an equal chance of being selected. This method is unbiased and ensures that each person has an equal probability of being chosen for the sample.

- **Example**: Suppose you want to conduct a survey in a city. You randomly select 20 people from the total population, giving each person the same chance to be part of the survey.
- **Result**: The plot shows the population with the randomly selected individuals highlighted in red.

## **2. Stratified Sampling**
In stratified sampling, the population is divided into subgroups (strata) based on specific characteristics, such as age, income level, or education. A random sample is taken from each subgroup to ensure all categories are represented proportionally.

- **Example**: You want to survey individuals from different age groups in a city. The population is divided into 4 age-based strata, and an equal number of individuals are selected from each stratum.
- **Result**: The plot highlights individuals sampled from each age group in green, ensuring all strata are represented.

## **3. Systematic Sampling**
In systematic sampling, every nth member of the population is selected after a random starting point. This method is useful when you have a sequential list of individuals.

- **Example**: You decide to survey every 5th person in a queue at a shopping mall. After randomly choosing a starting point, you select every 5th individual for the survey.
- **Result**: The plot shows the population with every 5th individual highlighted in orange.

## **4. Cluster Sampling**
In cluster sampling, the population is divided into clusters, such as neighborhoods or geographical regions. Instead of sampling individuals, entire clusters are randomly selected, and all individuals within the chosen clusters are surveyed.

- **Example**: A city is divided into 5 neighborhoods (clusters). You randomly select 2 neighborhoods and survey everyone living in those areas.
- **Result**: The plot highlights two entire clusters (neighborhoods) sampled in purple, where all individuals within those neighborhoods are included.

## **Conclusion**
Each sampling technique has its unique advantages and applications, depending on the nature of the population and the goal of the survey or study. Understanding these methods is crucial for collecting representative data efficiently and minimizing bias.


````python
import numpy as np
import matplotlib.pyplot as plt

# Create a population of 100 individuals (representing a city's population, numbered 1-100)
population_size = 100
population = np.arange(1, population_size + 1)

# Random Sampling: Every member of the population has an equal chance of being selected.
def random_sampling(population, sample_size):
    return np.random.choice(population, sample_size, replace=False)

# Stratified Sampling: The population is divided into subgroups (strata), and random samples are taken from each stratum.
# Example: Population divided into age groups.
def stratified_sampling(population, strata_size, sample_size):
    strata = np.array_split(population, strata_size)
    samples = []
    for stratum in strata:
        samples.extend(np.random.choice(stratum, sample_size // strata_size, replace=False))
    return np.array(samples)

# Systematic Sampling: Every nth member of the population is selected after a random starting point.
# Example: Surveying every 5th person in a queue.
def systematic_sampling(population, sample_size, step):
    start = np.random.randint(0, step)
    return population[start::step][:sample_size]

# Cluster Sampling: The population is divided into clusters, and entire clusters are sampled.
# Example: A city divided into neighborhoods, and entire neighborhoods are sampled.
def cluster_sampling(population, cluster_size, num_clusters):
    clusters = np.array_split(population, cluster_size)
    sampled_clusters = np.random.choice(range(cluster_size), num_clusters, replace=False)
    sampled = np.concatenate([clusters[i] for i in sampled_clusters])
    return sampled

# Parameters for sampling
sample_size = 20
strata_size = 4  # Dividing population into 4 strata (e.g., age groups)
cluster_size = 5  # Dividing population into 5 clusters (e.g., neighborhoods)
num_clusters = 2  # Sampling 2 entire clusters (neighborhoods)
step = population_size // sample_size  # Step size for systematic sampling

# Generate samples
random_sample = random_sampling(population, sample_size)
stratified_sample = stratified_sampling(population, strata_size, sample_size)
systematic_sample = systematic_sampling(population, sample_size, step)
cluster_sample = cluster_sampling(population, cluster_size, num_clusters)

# Create subplots for visualizing each sampling technique
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Types of Sampling Techniques')

# Plot Random Sampling: Every person has an equal chance to be selected
axs[0, 0].scatter(population, np.ones_like(population), label="Population")
axs[0, 0].scatter(random_sample, np.ones_like(random_sample), color='red', label="Random Sample")
axs[0, 0].set_title('Random Sampling: Equal chance for everyone')
axs[0, 0].legend()

# Plot Stratified Sampling: Sample is drawn equally from different strata (e.g., age groups)
axs[0, 1].scatter(population, np.ones_like(population), label="Population")
axs[0, 1].scatter(stratified_sample, np.ones_like(stratified_sample), color='green', label="Stratified Sample")
axs[0, 1].set_title('Stratified Sampling: Equal samples from strata (age groups)')
axs[0, 1].legend()

# Plot Systematic Sampling: Every nth member is sampled
axs[1, 0].scatter(population, np.ones_like(population), label="Population")
axs[1, 0].scatter(systematic_sample, np.ones_like(systematic_sample), color='orange', label="Systematic Sample")
axs[1, 0].set_title('Systematic Sampling: Sample every nth member')
axs[1, 0].legend()

# Plot Cluster Sampling: Entire clusters (neighborhoods) are sampled
axs[1, 1].scatter(population, np.ones_like(population), label="Population")
axs[1, 1].scatter(cluster_sample, np.ones_like(cluster_sample), color='purple', label="Cluster Sample")
axs[1, 1].set_title('Cluster Sampling: Entire clusters (neighborhoods) sampled')
axs[1, 1].legend()

# Adjust plot layout and display
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

# **Bias in Sampling and the Central Limit Theorem**

This notebook demonstrates two key concepts in statistics:
1. **Bias in Sampling**: How non-representative samples can lead to inaccurate conclusions.
2. **Central Limit Theorem (CLT)**: How the distribution of sample means approaches a normal distribution as we take more random samples from the population.

## **1. Bias in Sampling**

In some cases, the sample chosen may not be representative of the entire population. This can happen when we sample only from a specific subgroup (e.g., a demographic with lower income or older age). Such **biased samples** can skew the results and lead to inaccurate conclusions.

### **Example of Biased Sampling**
- We create a population with a normal distribution (mean = 50, standard deviation = 10).
- A biased sample is taken by selecting only individuals from a specific subgroup (e.g., those with values below 40).
- This biased sample results in a skewed mean that does not represent the population accurately.

### **Effect of Bias**
- **Biased Sampling** leads to a mean that is not reflective of the true population mean.
- If conclusions are drawn from such biased samples, the results can be misleading.

## **2. Random Sampling and the Central Limit Theorem**

In contrast to biased sampling, **random sampling** ensures that every individual in the population has an equal probability of being selected. This reduces bias and provides a more accurate representation of the population.

### **Central Limit Theorem (CLT)**
The CLT states that the distribution of sample means approaches a normal distribution, even if the original population is not normally distributed, as long as:
1. The sample size is sufficiently large.
2. Multiple random samples are taken.

### **Example of Random Sampling and CLT**
- We take 1000 random samples from the population and calculate their means.
- The distribution of these sample means forms a normal distribution, illustrating the CLT.

## **Results**
- The mean of the biased sample is significantly different from the true population mean.
- The mean of the random sample is much closer to the population mean.
- As more random samples are taken, the distribution of sample means converges to a normal distribution.

### **Conclusions**
- **Biased samples** lead to inaccurate and skewed results.
- **Random sampling** helps minimize bias, ensuring that the sample is representative of the population.
- The **Central Limit Theorem** ensures that the distribution of sample means will be normal, regardless of the population's distribution, as long as enough random samples are taken.


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Simulating the population: A normally distributed population with a skewed subgroup
population_size = 10000
population = np.random.normal(loc=50, scale=10, size=population_size)

# Simulate a biased sample by only selecting from a specific range (e.g., lower-income or older demographic)
biased_sample_size = 100
biased_population = population[population < 40]  # Biased towards individuals with a mean below 40

# Simulate a random sample from the population
random_sample_size = 100
random_sample = np.random.choice(population, random_sample_size, replace=False)

# Take a large number of samples for random sampling to illustrate the sampling distribution
num_samples = 1000
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, random_sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot the population distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.histplot(population, kde=True, bins=30, color='blue')
plt.title("Entire Population Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot the biased sample
plt.subplot(1, 3, 2)
sns.histplot(biased_population, kde=True, bins=30, color='red')
plt.title("Biased Sample (Only from specific subgroup)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot the random sample means distribution
plt.subplot(1, 3, 3)
sns.histplot(sample_means, kde=True, bins=30, color='green')
plt.title("Random Sampling Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Print out the means to compare the effects of bias
print(f"Mean of the entire population: {np.mean(population):.2f}")
print(f"Mean of the biased sample: {np.mean(biased_population):.2f}")
print(f"Mean of the random sample: {np.mean(random_sample):.2f}")
print(f"Mean of the sample means (random sampling): {np.mean(sample_means):.2f}")
````

# **Point Estimation and Confidence Intervals**

In this section, we will explore the concepts of **Point Estimation** and **Confidence Intervals**, which are fundamental to inferential statistics. These tools allow us to make educated guesses about a population based on sample data.

## **Problem Description**

We often cannot measure an entire population, so we take a sample and use **point estimates** to approximate population parameters (e.g., mean, variance). However, point estimates alone do not give us information about how close the estimate is to the true population parameter.

To account for this uncertainty, we calculate **confidence intervals**, which provide a range of values that is likely to contain the population parameter with a certain level of confidence (e.g., 95%).

### **Scenario**:
We have a sample of 100 data points drawn from a normally distributed population with a mean of 50 and a standard deviation of 10. We aim to:
1. Estimate the population mean using the sample mean (**Point Estimation**).
2. Calculate a **95% Confidence Interval** for the population mean to provide a range where the true mean is likely to lie.

### **Goals**:
- **Point Estimation**: Calculate the sample mean and use it as an estimate of the population mean.
- **Confidence Interval**: Provide a range (with 95% confidence) that is likely to contain the true population mean.

## **Interpretation of Results**:

1. **Point Estimate (Sample Mean)**: This value is a single number that serves as the best estimate for the population mean based on our sample. It gives us an idea of where the true population mean is likely to be, but it doesn't tell us how precise this estimate is.
   
2. **Confidence Interval**: A confidence interval gives us a range of plausible values for the population mean. In this case, a **95% Confidence Interval** means that if we were to take many samples and calculate a confidence interval for each, we expect that 95% of those intervals would contain the true population mean.
   - **Interpretation**: If our confidence interval is from 47.80 to 51.36, we can say with 95% confidence that the true population mean lies somewhere within this range.

The following code demonstrates how to compute both the **point estimate** and the **confidence interval** a


````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate a sample data from a normally distributed population
np.random.seed(42)
population_mean = 50
population_std = 10
sample_size = 100
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Point Estimation: Mean of the sample
point_estimate = np.mean(sample_data)
print(f"Point Estimate (Sample Mean): {point_estimate:.2f}")

# Confidence Interval: 95% Confidence Interval for the population mean
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)
confidence_interval = stats.t.interval(confidence_level, degrees_of_freedom, loc=point_estimate, scale=standard_error)

print(f"95% Confidence Interval: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}")

# Plotting the Sample Data Distribution and Confidence Interval
plt.figure(figsize=(10, 6))

# Plot histogram of the sample data
plt.hist(sample_data, bins=20, alpha=0.7, color='blue', edgecolor='black', density=True)

# Overlay the true population mean and sample mean
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='True Population Mean')
plt.axvline(point_estimate, color='green', linestyle='dashed', linewidth=2, label='Sample Mean (Point Estimate)')

# Plot the confidence interval
plt.axvline(confidence_interval[0], color='orange', linestyle='dashed', linewidth=2, label='95% Confidence Interval')
plt.axvline(confidence_interval[1], color='orange', linestyle='dashed', linewidth=2)

# Plot settings
plt.title("Point Estimation and Confidence Interval for Sample Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(loc='upper right')
plt.show()
````

# **Hypothesis Testing: Comparing Average Heights Between Two Groups**

In this section, we will perform a **hypothesis test** to compare the average height between two independent groups using a **two-sample t-test**.

## **Problem Description**:

Suppose we are interested in determining whether there is a significant difference in the average height between two different groups of individuals. For example, we may have two groups of people and want to test if Group 1 and Group 2 have different average heights.

### **Hypotheses**:
- **Null Hypothesis (H)**: There is no difference in average height between the two groups.
  - This means any observed difference in heights is due to random chance, not a true difference.
  
- **Alternative Hypothesis (H)**: There is a difference in average height between the two groups.
  - This means the difference in heights is statistically significant and not due to random chance.

### **Significance Level ()**:
- We will use a **significance level ()** of 0.05, which is a commonly accepted threshold.
  - If the **p-value** from the test is **less than 0.05**, we will **reject the null hypothesis** and conclude that there is a significant difference in average heights between the groups.
  - If the **p-value** is **greater than or equal to 0.05**, we will **fail to reject the null hypothesis** and conclude that there is no significant difference between the groups.

## **Steps**:
1. **Simulate Data**: We will simulate the heights of two groups of individuals.
   - **Group 1**: Sample of 30 individuals with an average height of 170 cm.
   - **Group 2**: Sample of 30 individuals with an average height of 175 cm.
   
2. **Perform Hypothesis Testing**: Using a **two-sample t-test**, we will test whether the difference in the average heights between the two groups is statistically significant.

3. **Interpret Results**:
   - **T-statistic**: The t-test statistic tells us how far apart the groups are in standard error units.
   - **P-value**: The p-value tells us the probability of observing a difference as extreme as this if the null hypothesis is true.

## **Interpretation of Results**:
- **Reject H**: If the p-value is less than 0.05, we will reject the null hypothesis and conclude that there is a statistically significant difference in the average heights between the two groups.
  
- **Fail to Reject H**: If the p-value is greater than or equal to 0.05, we will fail to reject the null hypothesis and conclude that there is no statistically significant difference in average heights between the two groups.

The following code demonstrates the hypothesis testing process and helps us understand the results with a visual representation of the height distributions for both groups.


````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulate the heights of two groups (in cm)
# Group 1: Sample of 30 individuals with mean 170 and std 10
group1_heights = np.random.normal(170, 10, 30)

# Group 2: Sample of 30 individuals with mean 175 and std 10
group2_heights = np.random.normal(175, 10, 30)

# Hypotheses:
# Null Hypothesis (H): There is no difference in average height between the two groups.
# Alternative Hypothesis (H): There is a difference in average height between the two groups.

# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group1_heights, group2_heights)

# Set significance level (alpha)
alpha = 0.05

# Print results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.5f}")

# Interpretation of the p-value
if p_value < alpha:
    print("Reject the Null Hypothesis (H): There is a significant difference in average height between the two groups.")
else:
    print("Fail to Reject the Null Hypothesis (H): There is no significant difference in average height between the two groups.")

# Visualizing the distributions of the two groups
plt.figure(figsize=(10, 6))

# Plot histograms of both groups
plt.hist(group1_heights, bins=10, alpha=0.5, label='Group 1 (mean=170)', color='blue')
plt.hist(group2_heights, bins=10, alpha=0.5, label='Group 2 (mean=175)', color='green')

# Plot settings
plt.axvline(np.mean(group1_heights), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(group2_heights), color='green', linestyle='dashed', linewidth=2)

plt.title('Height Distribution of Two Groups')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
````

# **Types of Errors in Hypothesis Testing**

In hypothesis testing, there are two types of errors that can occur, and understanding these errors is crucial for interpreting the results of a test:

## **Type I Error (False Positive)**:
- **Definition**: Rejecting the null hypothesis (**H**) when it is actually true.
- **Example**: Convicting an innocent person.
- **Probability**: This error occurs with probability **** (the significance level, typically 0.05).
  
  - **Illustration**: Suppose there is no real difference between two groups (H is true), but after performing a hypothesis test, we mistakenly reject H, believing there is a difference when none exists.

## **Type II Error (False Negative)**:
- **Definition**: Failing to reject the null hypothesis (**H**) when it is actually false (i.e., when the alternative hypothesis, **H**, is true).
- **Example**: Letting a guilty person go free.
- **Probability**: This error occurs with probability ****.
  
  - **Illustration**: Suppose there is a real difference between two groups (H is true), but the hypothesis test fails to detect it, and we mistakenly accept H.

### **2x2 Table for Type I and Type II Errors**

| **Actual Reality**         | **Decision: Reject H**                    | **Decision: Fail to Reject H**        |
|----------------------------|--------------------------------------------|----------------------------------------|
| **H is true (No difference)**  | **Type I Error (False Positive)**: We reject H, but H is true. | Correct: We correctly fail to reject H. |
| **H is true (Difference)**     | Correct: We correctly reject H.        | **Type II Error (False Negative)**: We fail to reject H, but H is true. |

### **Goal**:
In this section, we will simulate two groups of data:
- **Group 1**: Represents the true mean of the population.
- **Group 2 (Null Hypothesis)**: Has the same mean as Group 1 (H is true).
- **Group 2 (Alternative Hypothesis)**: Has a different mean from Group 1 (H is true).

We will perform hypothesis testing under both scenarios and determine whether **Type I** and **Type II** errors occur. We will also display the results in a **2x2 confusion matrix** to illustrate the outcomes.

The following code demonstrates this simulation and illustrates the concept of Type I and Type II errors in hypothesis testing.


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Set random seed for reproducibility
np.random.seed(42)

# Simulate data for two groups
# Group 1: Mean = 100, Std = 10 (true mean)
# Group 2: Mean = 100 for Null Hypothesis (same as Group 1)
# Group 2: Mean = 110 for Alternative Hypothesis (slightly different from Group 1)
group1 = np.random.normal(100, 10, 100)
group2_null = np.random.normal(100, 10, 100)  # Same mean (H is true)
group2_alt = np.random.normal(110, 10, 100)   # Different mean (H is true)

# Perform hypothesis tests
# H is true (no real difference)
t_stat_null, p_value_null = ttest_ind(group1, group2_null)

# H is true (there is a real difference)
t_stat_alt, p_value_alt = ttest_ind(group1, group2_alt)

# Set significance level ()
alpha = 0.05

# Determine Type I and Type II errors
type_i_error = p_value_null < alpha  # Reject H when it's actually true (false positive)
type_ii_error = p_value_alt >= alpha  # Fail to reject H when H is true (false negative)

# Create a 2x2 matrix for visualizing the outcomes
outcome_matrix = np.array([
    ['True Negative (Correct)', 'Type I Error (False Positive)'],
    ['Type II Error (False Negative)', 'True Positive (Correct)']
])

# Display the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(np.array([[1, int(type_i_error)], [int(type_ii_error), 1]]), annot=outcome_matrix, fmt='', cmap='Blues', cbar=False, linewidths=1, linecolor='black', square=True)
plt.title('Type I and Type II Errors: 2x2 Matrix')
plt.show()

# Print results
print(f"T-statistic when H is true: {t_stat_null:.2f}, P-value: {p_value_null:.5f}")
print(f"T-statistic when H is true: {t_stat_alt:.2f}, P-value: {p_value_alt:.5f}")

# Interpretation of errors
if type_i_error:
    print("Type I Error occurred: Rejected H when it was true (false positive).")
else:
    print("Correctly failed to reject H when it was true.")

if type_ii_error:
    print("Type II Error occurred: Failed to reject H when it was false (false negative).")
else:
    print("Correctly rejected H when H was true.")
````

# **Parametric Methods in Statistics**

Parametric methods are statistical techniques that rely on assumptions about the distribution of the data. These methods are widely used in hypothesis testing when we assume that the underlying population follows a specific distribution, typically a **normal distribution**.

## **Definition**:
- **Parametric methods** assume that the data follows a certain distribution (e.g., normal distribution). These methods are powerful and efficient when the assumptions are met.

## **Common Parametric Tests**:

1. **t-tests**: Compares the means of two groups. It is used when we want to determine whether there is a significant difference between the means of two independent groups.

2. **ANOVA (Analysis of Variance)**: Compares the means of three or more groups. It is used when we want to test whether the means of multiple groups are significantly different from each other.

3. **Chi-Square Tests**: Used for categorical data to assess how likely it is that any observed difference between categories occurred by chance. It compares the observed frequencies of events to the expected frequencies.

---

## **Example 1: Independent t-Test**
- **Goal**: Compare the means of two independent groups (e.g., two groups of students who received different study materials) to see if their average performance is significantly different.
- **Hypotheses**:
  - **Null Hypothesis (H)**: There is no difference in the means between the two groups.
  - **Alternative Hypothesis (H)**: There is a significant difference in the means between the two groups.

---

## **Example 2: One-Way ANOVA**
- **Goal**: Compare the means of three independent groups (e.g., testing the effectiveness of three different diets on weight loss) to see if their average outcomes are significantly different.
- **Hypotheses**:
  - **Null Hypothesis (H)**: All group means are equal.
  - **Alternative Hypothesis (H)**: At least one group mean is different.

---

## **Example 3: Chi-Square Test**
- **Goal**: Test the relationship between two categorical variables (e.g., gender and preference for a particular product) to determine if the observed frequencies differ from what is expected by chance.
- **Hypotheses**:
  - **Null Hypothesis (H)**: The observed frequencies are equal to the expected frequencies (no association between variables).
  - **Alternative Hypothesis (H)**: The observed frequencies differ from the expected frequencies (there is an association).

The following Python code demonstrates how to perform these tests and visualize the results using simulated data.


````python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Example 1: Independent t-test (comparing the means of two groups)
# Group 1: Mean = 70, Std = 10
group1 = np.random.normal(70, 10, 30)

# Group 2: Mean = 75, Std = 10
group2 = np.random.normal(75, 10, 30)

# Perform an independent t-test
t_stat, p_value_ttest = stats.ttest_ind(group1, group2)
print(f"t-Test Results: T-statistic = {t_stat:.2f}, P-value = {p_value_ttest:.5f}")

# Example 2: One-Way ANOVA (comparing the means of three groups)
# Group A: Mean = 60, Std = 10
group_a = np.random.normal(60, 10, 30)

# Group B: Mean = 65, Std = 10
group_b = np.random.normal(65, 10, 30)

# Group C: Mean = 70, Std = 10
group_c = np.random.normal(70, 10, 30)

# Perform one-way ANOVA
f_stat, p_value_anova = stats.f_oneway(group_a, group_b, group_c)
print(f"ANOVA Results: F-statistic = {f_stat:.2f}, P-value = {p_value_anova:.5f}")

# Example 3: Chi-Square Test (for categorical data)
# Observed data: [10, 20] (Category 1) and [30, 40] (Category 2)
observed = np.array([[10, 20], [30, 40]])

# Perform Chi-Square Test
chi2_stat, p_value_chi2, dof, expected = stats.chi2_contingency(observed)
print(f"Chi-Square Test Results: Chi-Square Statistic = {chi2_stat:.2f}, P-value = {p_value_chi2:.5f}")

# Displaying the expected frequencies for Chi-Square test
print("Expected Frequencies:\n", expected)

# Visualizing t-test and ANOVA distributions
plt.figure(figsize=(10, 6))

# Plot histograms for t-test groups
plt.subplot(1, 2, 1)
sns.histplot(group1, bins=10, color='blue', label='Group 1 (Mean = 70)', kde=True)
sns.histplot(group2, bins=10, color='green', label='Group 2 (Mean = 75)', kde=True)
plt.title("t-Test: Distribution of Two Groups")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Plot histograms for ANOVA groups
plt.subplot(1, 2, 2)
sns.histplot(group_a, bins=10, color='red', label='Group A (Mean = 60)', kde=True)
sns.histplot(group_b, bins=10, color='orange', label='Group B (Mean = 65)', kde=True)
sns.histplot(group_c, bins=10, color='purple', label='Group C (Mean = 70)', kde=True)
plt.title("ANOVA: Distribution of Three Groups")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
````
